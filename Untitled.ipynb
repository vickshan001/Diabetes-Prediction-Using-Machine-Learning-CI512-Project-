```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from sklearn.metrics import accuracy_score

df = pd.read_csv('diabetes.csv')
#Print the first 5 rows of the dataframe.
df.head()
```

```python
print('Train columns with null values:\n ', df.isnull().sum())
```

```python
#Let's observe the shape of the dataframe.
df.shape
```

```python
#Let's create numpy arrays for features and target
X = df.drop('Outcome',axis=1).values
y = df['Outcome'].values
```

```python
#importing train_test_split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.4,random_state=42, stratify=y)
```

```python
print("1.Number of the rows having zero's in glucose feature: {0}".format(len(df.loc[df['Glucose'] == 0])))
print("2.Number of the rows having zero's in bloodpressure feature: {0}".format(len(df.loc[df['BloodPressure'] == 0])))
print("3.Number of the rows having zero's in skinthickness feature: {0}".format(len(df.loc[df['SkinThickness'] == 0])))
print("4.Number of the rows having zero's in  insulin feature: {0}".format(len(df.loc[df['Insulin'] == 0])))
print("5.Number of the rows having zero's in  bmi feature: {0}".format(len(df.loc[df['BMI'] == 0])))

df['Glucose']=df['Glucose'].replace(0,df['Glucose'].mean())
df['BloodPressure']=df['BloodPressure'].replace(0,df['BloodPressure'].mean())
df['SkinThickness']=df['SkinThickness'].replace(0,df['SkinThickness'].mean())
df['Insulin']=df['Insulin'].replace(0,df['Insulin'].mean())
df['BMI']=df['BMI'].replace(0,df['BMI'].median())
```

```python
knnarray = []
dtarray = []
rfarray = []
mlparray = []
stackingarray = []

#import KNeighborsClassifier
neighbors = np.arange(1,9)
for i,k in enumerate(neighbors):
    from sklearn.neighbors import KNeighborsClassifier
    #Setup a knn classifier with k neighbors
    knn = KNeighborsClassifier(n_neighbors=k)
    #Fit the model
    knn.fit(X_train,y_train)
    #Get accuracy. Note: In case of classification algorithms score method represents accuracy.
    knn.score(X_test,y_test)
    knn = knn.fit(X_train, y_train)
    y_prediction = knn.predict(X_test)
    print("Train/test accuracy:", accuracy_score(y_test,y_prediction))
    knnarray.append(accuracy_score(y_test,y_prediction))
    
    ### RANDOM FOREST ##
    from sklearn.ensemble import RandomForestClassifier
    clfrandomforest = RandomForestClassifier()
    clfrandomforest = clfrandomforest.fit(X_train, y_train)
    randomforest_prediction = clfrandomforest.predict(X_test)
    print("Train/test accuracy: Random Forest",accuracy_score(y_test,randomforest_prediction))
    rfarray.append(accuracy_score(y_test,randomforest_prediction))
    ###

     ## DECISION TREE ##
    from sklearn.tree import DecisionTreeClassifier
    clfdecisiontree= DecisionTreeClassifier()
    clfdecisiontree = clfdecisiontree.fit(X_train, y_train)
    Ydecisiontree_prediction = clfdecisiontree.predict(X_test)
    print("Train/test accuracy: Decision Tree",accuracy_score(y_test,Ydecisiontree_prediction))
    dtarray.append(accuracy_score(y_test,Ydecisiontree_prediction))
    ## END OF DECISION TREE ##
    
    ## MLP CLASSIFIER ##
    from sklearn.neural_network import MLPClassifier
    MLPclf = MLPClassifier(random_state=1, max_iter= 1000).fit(X_train, y_train)
    MLPclf = knn.fit(X_train, y_train)
    MLPprediction = MLPclf.predict(X_test)
    print("Multi-Layer Perception =", accuracy_score(y_test,MLPprediction))
    mlparray.append(accuracy_score(y_test,MLPprediction))
    
    ### END OF MLP CLASSIFER ##
    
    ### STACKING CLASSIFER ###
    from sklearn.ensemble import StackingClassifier
    clfstacking = StackingClassifier(estimators = [('knn',knn),('dtc',clfdecisiontree), ('rfc', clfrandomforest)],final_estimator = clfrandomforest)
    clfstacking.fit(X_train, y_train)
    Ystacking_prediction = clfstacking.predict(X_test)
    print("Train/test Stacking accuracy", accuracy_score(y_test, Ystacking_prediction))
    stackingarray.append(accuracy_score(y_test, Ystacking_prediction))
    ## END OF STACKING ###
    
### CROSS FOLD VALIDATION ###

from sklearn.model_selection import ShuffleSplit
cv = ShuffleSplit(n_splits=10, test_size=0.7)

from sklearn.model_selection import cross_val_score
scores = cross_val_score(clfrandomforest, X, y, cv=cv)
print()
print("Cross fold validation accuracy scores:",scores)
print("Cross fold validation accuracy mean:",scores.mean())
    
    
```

```python
# loop through each algorithm and create a line graph
import matplotlib.pyplot as plt
plt.plot(knnarray, label='KNN')
plt.plot(dtarray, label='Decision Tree')
plt.plot (rfarray, label='Random Forest')
plt.plot (mlparray, label='MLP')
plt.plot (stackingarray, label='Stacking')
plt.legend()
plt.show()
```
